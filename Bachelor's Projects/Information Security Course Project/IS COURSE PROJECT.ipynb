```python
import hashlib as hsh
import time
import sys
import random
import string
import random 
import matplotlib.pyplot as plt
from prettytable import PrettyTable
from tabulate import tabulate
from timeit import default_timer as timer
from pympler.asizeof import asizeof
import timeit
```

```python
#timer
start_time = time.time()
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
#to get size of sha object
def getsize(z):
   return asizeof(z)
```

```python
#return size of digest or string
def utf8len(s):
    return len(s.encode('utf-8'))
```

```python
#random string generator
def randomString(stringLength):
    """Generate a random string with the combination of lowercase and uppercase letters """

    letters = string.ascii_letters
    return ''.join(random.choice(letters) for i in range(stringLength))
```

```python
#create list of random strings 
def random_list(y,start,end):
    rands=[]
    for i in range(0,y):
        r1=random.randint(start,end)
        rands.append(randomString(r1))
    return rands
```

```python
#function to return the digest size,block size,digest,hexadecimal value of digest
def SHA_1(x,y="abcd"):
    
    x.update(y.encode())
    print(x.digest_size)
    print(x.block_size)
    digest=x.digest()
    hexd=x.hexdigest()
    print(digest)
    print(hexd)
    sha1=[x.digest_size,x.block_size,digest,hexd]
    return sha1
```

```python

def SHA_256(x,y="abcd"):
    
    x.update(y.encode())
    print(x.digest_size)
    print(x.block_size)
    digest=x.digest()
    hexd=x.hexdigest()
    print(digest)
    print(hexd)
    sha256=[x.digest_size,x.block_size,digest,hexd]
    return sha256
```

```python
def SHA_512(x,y="abcd"):
    
    x.update(y.encode())
    print(x.digest_size)
    print(x.block_size)
    digest=x.digest()
    hexd=x.hexdigest()
    print(digest)
    print(hexd)
    sha512=[x.digest_size,x.block_size,digest,hexd]
    return sha512
```

```python
#get random list to caluclate the variables
randoms=random_list(100,100,150)
randoms
```

```python
# to calculate the list which contains all the required evaluation variables of SHA-1

x=[]
start = timer()

for i in range(0,100):
    x.append(hsh.sha1())
sha_list=[]
for i in range(0,100):
    
    str1=randoms[i]
    temp=SHA_1(x[i],str1)
    temp.append(utf8len(str1))
    temp.append(sys.getsizeof(str1))
    sha_list.append(temp)

print(sha_list)
end = timer() 
print(end - start)
alg_time={'sha1':end - start}
```

```python
# to calculate the list which contains all the required evaluation variables of SHA-256
x256=[]
start = timer()
for i in range(0,100):
    x256.append(hsh.sha256())
sha256_list=[]
for i in range(0,100):
    
    str1=randoms[i]
    temp=SHA_256(x256[i],str1)
    temp.append(utf8len(str1))
    temp.append(sys.getsizeof(str1))
    sha256_list.append(temp)

print(sha256_list)
end = timer() 
print(end - start)
alg_time["sha256"]=end - start
```

```python

# to calculate the list which contains all the required evaluation variables of SHA-512
x512=[]
start = timer()
for i in range(0,100):
    x512.append(hsh.sha512())
sha512_list=[]
for i in range(0,100):
    
    str1=randoms[i]
    temp=SHA_512(x512[i],str1)
    temp.append(utf8len(str1))
    temp.append(sys.getsizeof(str1))
    sha512_list.append(temp)

print(sha512_list)
end = timer() 
print(end - start)
alg_time["sha512"]=end - start
```
